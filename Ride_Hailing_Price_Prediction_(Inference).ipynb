# Model Inference

Model yang sudah dilatih akan dicoba pada data yang bukan termasuk ke dalam train-set ataupun test-set.

# Import Library

```python
import pickle
import json
import pandas as pd
import numpy as np
```

# Load Dataset

```python
with open('model_lin.pkl', 'rb') as file_1:
    model_lin = pickle.load(file_1)

with open('model_scaler.pkl', 'rb') as file_2:
    model_scaler = pickle.load(file_2)

with open('model_encoder_ordinal.pkl', 'rb') as file_3:
    model_encoder_ordinal = pickle.load(file_3)

with open('model_encoder_nominal.pkl', 'rb') as file_4:
     model_encoder_nominal = pickle.load(file_4)

with open('list_num_col.txt', 'rb') as file_5:
    list_num_column = pickle.load(file_5)

with open('list_cat_col.txt', 'rb') as file_6:
    list_cat_cal = pickle.load(file_6)

with open('list_cat_col_nominal.txt', 'rb') as file_7:
    list_cat_cal_n = pickle.load(file_7)

with open('list_cat_col_ordinal.txt', 'rb') as file_8:
    list_cat_cal_o = pickle.load(file_8)
```

# Create Data

```python
data_inf = {'name': {0: 'UberX',
                        1: 'Lux',
                        2: 'Lyft',
                        3: 'Lux Black XL',
                        4: 'Lyft XL'},
            'short_summary': {0: 'Overcast',
                        1: 'Rain',
                        2: 'Clear',
                        3: 'Clear',
                        4: 'Partly Cloudy'},
            'hour': {0: 10, 1: 2, 2: 1, 3: 4, 4: 3},
            'day': {0: 21, 1: 27, 2: 28, 3: 30, 4: 29},
            'distance': {0: 1, 1: 0.44, 2: 0.44, 3: 0.44, 4: 0.44},
            'cab_type': {0: 'Uber', 1: 'Lyft', 2: 'Lyft', 3: 'Lyft', 4: 'Lyft'},
            'surge_multiplier': {0: 'high',
                                 1: 'high',
                                 2: 'high',
                                 3: 'high',
                                 4: 'high'}}
```

```python
data_inf = pd.DataFrame(data_inf)
data_inf
```

# Inferencing

```python
data_inf_num = data_inf[list_num_column]
data_inf_cat_ordinal = data_inf[list_cat_cal_o]
data_inf_cat_nominal = data_inf[list_cat_cal_n]
```

```python
data_inf_num_scaled = model_scaler.transform(data_inf_num)

data_inf_cat_ordinal_encoded = model_encoder_ordinal.transform(data_inf_cat_ordinal)

data_inf_cat_nominal_encoded = model_encoder_nominal.transform(data_inf_cat_nominal).toarray()
```

```python
data_inf_final = np.concatenate([data_inf_num_scaled, data_inf_cat_ordinal_encoded, data_inf_cat_nominal_encoded], axis=1)
```

```python
y_pred_inf = model_lin.predict(data_inf_final)
y_pred_inf
```

```python
data_inf['new_price'] = y_pred_inf
data_inf
```
